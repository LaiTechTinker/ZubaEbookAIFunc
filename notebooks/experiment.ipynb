```python
from langchain_community.document_loaders.pdf import PyPDFLoader
from langchain_community.document_loaders import Docx2txtLoader
```

```python
pdf_file_path="../books/CHE242.pdf"
word_file_path="../books/gns_111.docx"
```

```python
#pdf loader
loader = PyPDFLoader(pdf_file_path)
pdf_data = loader.load()
pdf_data
```

```python
#word_loader
loader = Docx2txtLoader(word_file_path)
docx_data = loader.load()
docx_data
```

```python
#pdf and docx splitting
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=200)
splitted_pdf = text_splitter.split_documents(pdf_data)
splitted_docx= text_splitter.split_documents(docx_data)
print(splitted_pdf[0])
print(splitted_docx[0])
```

```python
#this will upload to pinecone index
from langchain_pinecone import PineconeVectorStore
def ingest_book(book):
    vectorstore = PineconeVectorStore.from_documents(
        documents=book,
        embedding=model,
        index_name="zuba-aibook-index"
    )
    return vectorstore
```

```python
docx_vectorstore=ingest_book(splitted_docx)
pdf_vectorstore=ingest_book(splitted_pdf)
```

```python
#this is for retreiveing from pinecone 
def book_retrieve(user_id,book_id):
    doc_search=PineconeVectorStore.from_existing_index(
    index_name="zuba-aibook-index",
    embedding=model
)
    retrieval=doc_search.as_retriever(
        search_type="similarity",
        search_kwargs={
            "k":3,
            "filter":{
                "user_id":user_id,
                "book_id":book_id
            }
        }
    )
    return retrieval
```

```python
#now let's write some query for our retreival
user_query="does most closed system experience change in kinetic energy"
retreival=book_retrieve(user_id="user_123",book_id="book_456")
retreived_ans=retreival.invoke(user_query)
retreived_ans
```
